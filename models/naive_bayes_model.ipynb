```python
columns_to_remove = ["budget","homepage","original_language", "tagline", "tittle", "production_countries", "production_companies","release_date","revenue","runtime","spoken_languages","tagline"]

movies = big_table.drop(columns=columns_to_remove, axis=1)
movies
```

```python
import ast

def convert_genres(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert_genres)
```

```python
movies.head(2)
```

```python
def convert_cast(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter += 1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head(2)
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['Director'] = movies['crew'].apply(fetch_director)
movies.drop('crew',axis="columns",inplace=True)
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert_genres)
```

```python
def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['Director'] = movies['Director'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
def make_str(text):
    st = ""
    for i in text:
        st = st+i+" "
    return st

movies['keywords'] = movies['keywords'].apply(make_str)
movies['genres'] = movies['genres'].apply(make_str)
movies['cast'] = movies['cast'].apply(make_str)
movies['Director'] = movies['Director'].apply(make_str)
movies.head(2)
```

```python
print(movies.columns)
```

```python
vote_average_threshold = 7.0
vote_count_threshold = 1000
popularity_threshold = 50.0

# Create the "is_popular" column based on the criteria
movies['is_popular'] = (
    (movies['vote_average'] >= vote_average_threshold) &
    (movies['vote_count'] >= vote_count_threshold) &
    (movies['popularity'] >= popularity_threshold)
).astype(int)
```

```python
movies['tags'] = movies['overview']+movies['keywords']+movies['genres']+movies['cast']+movies['Director']
movies['tags'] = movies['tags'].astype(str)
```

```python
new_df=movies[['id','title','tags','is_popular']]
```

```python
def make_into_lower(text):
    st=""
    for i in text:
        st = st + i.lower()
    return st

movies['tags'] = movies['tags'].apply(make_into_lower)
```

```python
movies
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
movies["tags"] = movies["tags"].apply(stem)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(movies.tags,movies.is_popular)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer(max_features=5000 , stop_words = 'english')
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
('vectorizer', CountVectorizer()),
('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, roc_curve, auc, precision_recall_curve
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.datasets import fetch_20newsgroups
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix
plt.imshow(conf_matrix, interpolation='nearest')
plt.title('Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# ROC Curve
y_probs = model.predict_proba(X_test_count)
fpr, tpr, thresholds = roc_curve(y_test, y_probs[:, 1])
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
# Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_probs[:, 1])

# Plot Precision-Recall Curve
plt.figure()
plt.plot(recall, precision, color='darkorange', lw=2, label='Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower right')
plt.show()
```
