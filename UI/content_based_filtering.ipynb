```python
columns_to_remove = ["budget","homepage","original_language", "tagline", "tittle", "production_countries", "production_companies","release_date","revenue","runtime","spoken_languages","tagline"]

movies = big_table.drop(columns=columns_to_remove, axis=1)
movies
```

```python
import ast

def convert_genres(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert_genres)
```

```python
movies.head(2)
```

```python
def convert_cast(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter += 1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head(2)
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['Director'] = movies['crew'].apply(fetch_director)
movies.drop('crew',axis="columns",inplace=True)
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert_genres)
```

```python
def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['Director'] = movies['Director'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
def make_str(text):
    st = ""
    for i in text:
        st = st+i+" "
    return st

movies['keywords'] = movies['keywords'].apply(make_str)
movies['genres'] = movies['genres'].apply(make_str)
movies['cast'] = movies['cast'].apply(make_str)
movies['Director'] = movies['Director'].apply(make_str)
movies.head(2)
```

```python
print(movies.columns)
```

```python
movies['tags'] = movies['overview']+movies['keywords']+movies['genres']+movies['cast']+movies['Director']
movies['tags'] = movies['tags'].astype(str)
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
#new_df.iloc[0]['tags']
```

```python
def make_into_lower(text):
    st=""
    for i in text:
        st = st + i.lower()
    return st

movies['tags'] = movies['tags'].apply(make_into_lower)
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
movies["tags"] = movies["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words = 'english')

vector = cv.fit_transform(movies['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity_array = []
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        similarity_array.append(i[1])
        print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight')
recommend("Pirates of the Caribbean: At World's End")
#recommend("Spider-Man")
```
